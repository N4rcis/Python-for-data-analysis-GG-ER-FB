**Here we are going the repeat the process but by changing the values in the dictionnaries we'll create a dataframe adapted for the Modeling part**

```python
# Creation of the dataframe that will be used for the modeling part
# First the dataframe is a copy of the age column of the classical dataframe
df_m = df_v.Age

# We collect the columns list for later purpose
columns = df_v.columns

# We get the binary labelisation of the age column 
# Then we remove the last column of the now binary labelized category
age_d = pd.get_dummies(df_v.Age)
age_d.drop([age_d.columns[len(age_d.columns)-1]], axis=1, inplace=True)

# We join the new category to the old nominal one and drop the later
df_m = pd.concat([df_m, age_d], axis=1, join='inner')
df_m.drop(["Age"], axis=1, inplace=True)


# We now repeat the process for each nominal category following the columns order
for i in range(1, 5):
    col_d = pd.get_dummies(df_v[columns[i]])
    col_d.drop([col_d.columns[len(col_d.columns)-1]], axis=1, inplace=True)
    df_m = pd.concat([df_m, col_d], axis=1, join='inner')

# We add the non-nominal columns to the modeling dataframe without applying the process 
df_m = pd.concat([df_m, df_v[["Nscore", "Escore", "Oscore", "Ascore", "Cscore", "Impulsive", "SS"]]], axis=1, join='inner')

# We apply the process to the last categories (drugs consumption frequency)
for i in range(12, len(columns)):
    col_d = pd.get_dummies(df_v[columns[i]])
    col_d.drop([col_d.columns[len(col_d.columns)-1]], axis=1, inplace=True)
    for j in range(0, len(col_d.columns)):
        col_d.rename(columns={col_d.columns[j]:(columns[i] + " " + col_d.columns[j])}, inplace=True)
    df_m = pd.concat([df_m, col_d], axis=1, join='inner')

# We assure ourselves that all the columns are excusively numerical
pd.set_option('display.max_rows', 145)
print(df_m.dtypes)
```

### Age and Drug Use: Analyze how age correlates with the use of different drugs.

```python

drugs = ['Alcohol', 'Amphet', 'Amyl', 'Benzos', 'Caff', 'Cannabis', 'Choc', 'Coke', 'Crack', 'Ecstasy', 'Heroin', 'Ketamine', 'Legalh', 'LSD', 'Meth', 'Mushrooms', 'Nicotine', 'Semer', 'VSA']
# for lisibility, we could only display 4 barplots :
#drugs = ['Alcohol', 'Cannabis', 'Coke', 'Nicotine']


df_usage_mapping = df.copy()

# Mapping for drug usage : 

usage_mapping = {
    'Never Used': 0,
    'Used over a Decade Ago': 1,
    'Used in Last Decade': 5,
    'Used in Last Year': 10,
    'Used in Last Month': 50,
    'Used in Last Week': 100,
    'Used in Last Day': 200
}

for drug in drugs:
    df_usage_mapping[f'{drug}'] = df_usage_mapping[drug].map(lambda x: usage_mapping.get(x, x))
    

# Barplot : 
age_groups = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
sns.set_palette('viridis',n_colors=1)
for age_group in age_groups:
    age_df = df_usage_mapping[df_usage_mapping['Age'] == age_group]
    drug_columns = [f'{drug}' for drug in drugs]
    drug_subset = age_df[drug_columns]
    drug_usage_scores = drug_subset.sum() # Calculate the score for each drug usage by summing the scores in the mapping
    drug_usage_scores.plot(kind='bar', title=f'Drug Usage for Age Group {age_group}')
    plt.xlabel('Drugs')
    plt.ylabel('Irrelevant Numbers')
    plt.show()
```

```python
def classify_disorder(n_score, e_score, o_score, a_score, c_score, 
                      n_high=42, e_high=44, o_high=51, a_high=48, c_high=46, 
                      n_low=29, e_low=35, o_low=41, a_low=39, c_low=37):
    disorders = []
    if n_score > n_high :
        if a_score < a_low or c_score < c_low:
            disorders.append('Anxiety')
            
    elif n_score > n_high or e_score < e_low:
        disorders.append('Depression')
        
    elif c_score > c_high :
        disorders.append('Panic')
            
    elif o_score > o_high and e_score > e_high:
        disorders.append('OCD')  
        
    else:
        disorders.append('No Disorder')

    return ', '.join(disorders)


# Apply the classification function to create a new column 'Disorder_Category' in the original DataFrame
df['Disorder_Category'] = df.apply(lambda row: classify_disorder(row['Nscore'], row['Escore'], row['Oscore'], row['Ascore'], row['Cscore']), axis=1)

# Replace cases where 'No Disorder' is not the only value with 'Not usable'
df['Disorder_Category'] = df['Disorder_Category'].apply(lambda x: 'Not Usable' if ',' in x and 'No Disorder' in x else x)

# Create a new DataFrame 'disorder_df' containing only individuals with a disorder
disorder_df = df[df['Disorder_Category'] != 'No Disorder'].copy()
disorder_df = df[df['Disorder_Category'] != 'Not usable'].copy()

# Insert the Disorder_Category column after the neoffi scores columns
insert_position = disorder_df.columns.get_loc('Cscore') + 1
disorder_df.insert(insert_position, 'Disorder_Category', disorder_df.pop('Disorder_Category'))
```

```python
#méthode de split des X,y avec processing MultiLabelBinarizer()
def mlb_split(df,split_index=33,test_size=0.2,random_state=random.randint(0, 100)):
    noms_colonnes = df.columns.tolist()
    name_variable_col=noms_colonnes[0:split_index]
    name_result_col=noms_colonnes[split_index:]
    
    X,y = df[name_variable_col].values, df[name_result_col].values
    
    #permet d'avoir au moins une instance de chaque y possible dans le train
    while True:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
        if not len(set(np.unique(y_train))) < len(set(np.unique(y))):
            break
        
    mlb = MultiLabelBinarizer()
    y_train_mlb = mlb.fit_transform(y_train)
    y_test_mlb = mlb.transform(y_test)
    
    return mlb,X_train, X_test, y_train_mlb, y_test_mlb
```

```python
#méthode d'affichage des métriques de performance des modèle pour une classification multi-label
def metrics(model,y_test_mlb, predictions, average='samples'):
    precision = precision_score(y_test_mlb, predictions, average=average)
    rappel = recall_score(y_test_mlb, predictions, average=average)
    f1 = f1_score(y_test_mlb, predictions, average=average)
    hamming_loss_value = hamming_loss(y_test_mlb, predictions)
    print('Precision_',model,': ',precision)
    print('Rappel_',model,': ',rappel)
    print('f1_',model,': ',f1)
    print('Hamming Loss_', model, ': ', hamming_loss_value)
    print('\n')
```

```python
#méthode d'application d'un modèle sur des données issu de mlb_split
def OVR_mlb(model,mlb,X_train,X_test,y_train_mlb,y_test_mlb):

    OVRmodel = OneVsRestClassifier(model)
    OVRmodel.fit(X_train, y_train_mlb)

    predictions = OVRmodel.predict(X_test)

    predicted_labels = mlb.inverse_transform(predictions)

    metrics(model,y_test_mlb, predictions)
```

```python
#méthode OVR_mlb + affichage des matrices de confusion des catégories qui sont le plus sources d'erreurs de prédiction
def confusion_mat(model,mlb,X_train,X_test,y_train_mlb,y_test_mlb):
    
    OVRmodel = OneVsRestClassifier(model)
    OVRmodel.fit(X_train, y_train_mlb)

    predictions = OVRmodel.predict(X_test)

    predicted_labels = mlb.inverse_transform(predictions)

    metrics(model,y_test_mlb, predictions)
    
    
    conf_matrix=multilabel_confusion_matrix(y_test_mlb, predictions)

    error_rates = np.array([(conf_matrix_class[0, 1] + conf_matrix_class[1, 0]) / conf_matrix_class.sum()
                        for conf_matrix_class in conf_matrix])

    high_error_indices = np.argsort(error_rates)[::-1]

    high_error_indices = high_error_indices[error_rates[high_error_indices] > 0.25]

    num_per_row = 5 
    subplot_size = math.ceil(len(high_error_indices) / num_per_row)

    plt.figure(figsize=(15, 4 * subplot_size))

    subplot_index = 1
    for i in high_error_indices:
        class_name = mlb.classes_[i]
        plt.subplot(subplot_size, num_per_row, subplot_index)
        subplot_index += 1
        sns.heatmap(conf_matrix[i], annot=True, fmt='d', cmap='Blues', cbar=False)
        plt.title(class_name)
        plt.xlabel("Predicted Label")
        plt.ylabel("True Label")

    plt.tight_layout()
    plt.show()
```
